```python
import numpy as np
import pandas as pd

display(np.random.randint(1,11,size = 10))  #注意：此方法不包含最大值
scores = np.random.randint(50,101,size = (50,5))
scores_pd = pd.DataFrame(scores,columns = ["國文","英文","數學","地理","歷史"],index = range(1,51))
scores_pd
```

```python
display(scores_pd.index)
display(scores_pd.columns)
scores_pd.index.name = "學號"
scores_pd.columns.name = "科目"
display(scores_pd)
```

```python
display(scores_pd.info())
display(scores_pd["國文"][:10])  #scores_pd["國文"]為一Series，然後抓取前10筆資料[:10]
display(scores_pd["數學"][:10])
display(scores_pd["國文"] + scores_pd["英文"] + scores_pd["數學"] + scores_pd["地理"] + scores_pd["歷史"])
```

```python
scores_pd["總分"] = scores_pd.sum(axis = 1)
scores_pd
```

```python
scores_pd["平均"] = scores_pd[["國文","英文","數學","地理","歷史"]].mean(axis = 1)
scores_pd
```
