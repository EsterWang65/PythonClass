```python
#陣列運算及統計
##亂數搭配陣列的運用
'''以前單純用random module來建立亂數只能建立一個，若是想建立多個亂數則可用numpy module中的numpy.random module來做。
random.randint(low, high=None, size=None, dtype=int): Return random integers from low (inclusive) to high (exclusive).
其中參數low和high可為int or array-like of ints；
而size可為int or tuple of ints，代表Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn.'''

import numpy as np
import pandas as pd

display(np.random.randint(1,11,size = 10))  #注意：此方法不包含最大值
scores = np.random.randint(50,101,size = (50,5))  #50列*5欄
scores_pd = pd.DataFrame(scores,columns = ["國文","英文","數學","地理","歷史"],index = range(1,51))
scores_pd
```

```python
##欄位調整處理
'''index和columns為objects，可透過該物件的property: index.name和columns.name來指定index和columns的名稱。'''

display(scores_pd.index)
display(scores_pd.columns)
scores_pd.index.name = "學號"
scores_pd.columns.name = "科目"
display(scores_pd)
```

```python
##陣列運算
'''若多個陣列間有相同的索引(如此處的索引皆為學號1~50)則可透過數學運算子做陣列運算。
注意：型態和數量要相符，因此事先可用info此method來查詢資料相關資訊。'''

display(scores_pd.info())
display(scores_pd["國文"][:10])  #scores_pd["國文"]為一Series，然後抓取前10筆資料[:10]
display(scores_pd["數學"][:10])
display(scores_pd["國文"] + scores_pd["英文"] + scores_pd["數學"] + scores_pd["地理"] + scores_pd["歷史"])
```

```python
##陣列的統計
'''可透過DataFrame中的method來做統計，如DataFrame物件名稱.sum()可作加法統計: Return the sum of the values over the requested axis.
可加入參數axis = index(0)/columns(1)來指定要針對index或columns來執行統計。'''

scores_pd["總分"] = scores_pd.sum(axis = 1)  #由左而右執行統計；新增欄位
scores_pd
```

```python
##陣列的統計
'''注意：因上面以新增總分欄位，若向上面直接對整個DataFrame進行統計運算，系統會將總分的值也一起加入運算中，
如此數值就會是錯誤的，因此此處的做法為使用subscript[]選取欲做統計之欄位，接著搭配統計method來執行。'''

scores_pd["平均"] = scores_pd[["國文","英文","數學","地理","歷史"]].mean(axis = 1)
scores_pd
```

```python
##資料排名
'''可透過rank此method來做資料排名，會回傳Series。DataFrame物件名稱.rank(): Compute numerical data ranks (1 through n) along axis.
By default, equal values are assigned a rank that is the average of the ranks of those values.
可加入參數axis來指定要針對index或columns進行排名；
加入參數ascending可指定排名為由大至小或由小至大，為bool，預設為True(小至大)，意思為value值小的排名小(靠前)；
加入參數method可指定相同內容值得資料要如何進行排名，可為'average', 'min', 'max', 'first'以及'dense'，預設為'average': 
How to rank the group of records that have the same value (i.e. ties):
"average": average rank of the group         "min": lowest rank in the group
"max": highest rank in the group             "first": ranks assigned in order they appear in the array
"dense": like 'min', but rank always increases by 1 between groups.'''

scores_pd["排名"] = scores_pd["總分"].rank(method = "min",ascending = False)  #value值由大至小排名；相同value取最小的rank值為rank
scores_pd
```

```python
##資料排序
scores_pd.sort_values(by = "排名")
```

```python
##DataFrame外觀處理
'''可透過DataFrame中的style此property來做DataFrame外觀(顯示)的處理，會傳回Styler object。
DataFrame物件名稱.style: Returns a Styler object. Contains methods for building a styled HTML representation of the DataFrame.
而Styler object中又有許多的property, method及attribute可使用。
如Styler.format(): Format the text display value of cells.，
可加入參數formatter指定value值要如何顯示(Object to define how values are displayed.)，可為str, callable(function), dict or None。
This method assigns a formatting function, formatter, to each cell in the DataFrame. If formatter is None, 
then the default formatter is used. If a callable then that function should take a data value as input and 
return a displayable representation, such as a string. If formatter is given as a string this is assumed to be 
a valid Python format specification and is wrapped to a callable as string.format(x). 
If a dict is given, keys should correspond to column names, and values should be string or callable, as above.；
加入參數subset可指定欲做此處理之資料，可為label, array-like, IndexSlice，
subset: A valid 2d input to DataFrame.loc[<subset>], or, in the case of a 1d input or single key, 
to DataFrame.loc[:, <subset>] where the columns are prioritised, to limit data to *before* applying the function.；
加入參數precision可指定小數點位數，注意只能為int。'''

def format60(v:int):
    if v >= 60:
        return "及格"
    else:
        return "不及格"

scores_pd[["國文","英文","數學","地理","歷史"]].style.format(format60)
```

```python
##DataFrame外觀處理
'''可透過highlight_between此method來用指定的樣式標記特定資料值，會傳回Styler。
Styler.highlight_between(): Highlight a defined range with a style.可加入參數subset，用法同上；
參數left代表Left bound for defining the range，可為scalar or datetime-like, or sequence or array-like，預設為None；
參數right代表Right bound for defining the range，可為scalar or datetime-like, or sequence or array-like，預設為None；
參數inclusive代表Identify whether bounds are closed or open，可為both', 'neither', 'left', 'right'；
參數color可指定標記的底色(background color)，預設為"yelllow"，注意只能為str；
參數props代表CSS properties to use for highlighting，注意只能為str，預設為None，
注意：若用了props此參數，則不能再用color參數了，否則會顯示錯誤。

小技巧：若有.的語法太長，可在.之前或之後加上反斜線\後即可換行，注意：只適用有.的程式段。
小技巧：若想對同一個Styler object進行多種method，可以直接將不同的method串接於前一個method後，
如此便可以不用用多行來處理同一個object了。'''

scores_pd.style\
.highlight_between(subset = ["國文","英文","數學","地理","歷史"],
                   left = 0,
                   right = 59,
                   props = "color:white;background-color:red")\
.format(subset = ["平均","排名"],precision = 0).to_excel("scores.xlsx")
```
