```python
import pandas as pd
from sqlalchemy import create_engine

df1 = pd.read_csv("world.csv")
taiwan2021 = df1.query('國家=="台灣"')
engine = create_engine("sqlite:///cov19.db")
taiwan2021.to_sql("台灣2021",engine,if_exists = "replace",index = False)
```

```python
import pandas as pd

df1 = pd.read_csv("world.csv")
usa = df1.query('國家=="美國"')
sortedFrame = usa.sort_values(by = "新增確診數",ascending = False).iloc[:100]
display(sortedFrame)
sortedFrame.to_json("USA_DATA.json",force_ascii = False,orient = "table")
```
