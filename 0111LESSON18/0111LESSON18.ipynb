```python
#資料處理
##讀檔
import pandas as pd

df1 = pd.read_csv("world.csv")
display(df1)
display(df1.info())
```

```python
##清理欄位(選取需要的資料)
df2 = df1.reindex(columns = ["洲名","國家","日期","總確診數","新增確診數","總人口數","解封指數"])  
#注意：該method會回傳一新DataFrame，故記得要用變數來接收
display(df2)
display(df2.info())
```

```python
##將「日期」欄位轉為日期型態
'''若資料中有日期欄位，一律建議將該欄位轉為日期型態。如此處日期原為字串型態，可透過pandas中的to_datetime此function來轉換型態，
注意：此方法會回傳一新Series(TimestampSeries)，故要記得用變數來接收。
pd.to_datetime(arg): Convert argument to datetime. This function converts a scalar, array-like, Series or
DataFrame/dict-like to a pandas datetime object.其中參數arg代表 The object to convert to a datetime. If a DataFrame is provided,
the method expects minimally the following columns: "year","month", "day". The column "year" must be specified in 4-digit format.
型態可為int, float, str, datetime, list, tuple, 1-d array,Series, DataFrame/dict-like。'''

df2["日期"] = pd.to_datetime(df2["日期"])
display(df2.info())
```

```python
##抓取資料
'''DataFrame搜尋value的方式有兩種，一種為使用subscript[]搭配條件運算子搜尋(subscript[[bool]])，另一種為使用query()搜尋。
DataFrame物件名稱.query(expr): a method that query the columns of a DataFrame with a boolean expression.
會回傳DataFrame或None(參數inplace = True)，其中參數expr(運算式)代表The query string to evaluate.注意：只能為str。
寫法：DataFrame物件名稱.query('欄位名稱+條件運算子+欲比對的值')，注意：「欲比對的值」若為str，則要再加上雙引號將其包裹。
使用query的好處為可在()透過邏輯運算子(and, or)加入多個條件，使用上比subscript靈活。'''

display(df2["洲名"] == "亞洲")  #使用subscript搭配條件運算子搜尋，會回傳bool的Series
display(df2[df2["洲名"] == "亞洲"])  #使用subscript搭條件運算子搜尋並抓取資料，會回傳DataFrame，注意：不會更改原DataFrame，僅抓取資料
display(df2.query('國家 == "台灣"'))  #使用query()搜尋並抓取資料
display(df2.query('國家 == "日本"'))  ##使用query()搜尋並抓取資料
```

```python
###多條件抓取資料
taiwan_japan = df2.query('國家=="台灣" or 國家=="日本"')  ##使用query()進行多條件搜尋並抓取資料
taiwan_japan
```

```python
##抓取資料並將DataFrame轉為excel檔
year2021 = taiwan_japan.query('日期>="2021-01-01" and 日期<"2022-01-01"')
display(year2021)
year2021.to_excel("year2021.xlsx")
```

```python
##檔案(上面存的excel檔)讀取，並透過參數usecols選取想要的資料欄位及index_col指定欲成為index之欄位
import pandas as pd

df3 = pd.read_excel("year2021.xlsx",usecols = ["國家","日期","總確診數","新增確診數"],index_col="國家")
display(df3)
display(df3.info())
```

```python
##資料排序
'''可透過sort_values此method來進行value的排序。DataFrame物件名稱.sort_values(by): Sort by the values along either axis.
會回傳DataFrame，其中參數代表 Name or list of names to sort by.可為str or list of str；
也可加入參數axis來指定要針對index/0或columns/1來排序，預設為0；
參數ascending可指定value排序方式為大至小或小至大，預設為True(小至大)，可為bool or list of bool。
資料排序後便可方便抓取資料最大或最小的前幾筆。'''

sortedFrame = df3.sort_values(by = "新增確診數",ascending = False)
display(sortedFrame)
sortedFrame.iloc[:20]  #抓取前20筆資料
```

```python
##將資料轉為資料庫sql檔案
'''一般要給上班族使用的話多為csv或excel檔，若是要給工程師運用則多會給json及資料庫。
在沒有雲端資料庫的情況下可使用個人資料庫sqlite，而sqlite實際上就是一個檔案。
可匯入sqlalchemy(The Python SQL Toolkit and Object Relational Mapper)此module來操作以上動作。
首先可透過sqlalchemy中的create_engine此function建立一engine以供後續存放資料及檔案使用。
(The Engine is the starting point for any SQLAlchemy application. It’s “home base” for the actual database and its DBAPI,
delivered to the SQLAlchemy application through a connection pool and a Dialect, which describes how to talk to 
a specific kind of database/DBAPI combination. The general structure can be illustrated as follows:
Database -- DBAPI -- Pool and Dialect -- Engine --> connect().
Where above, an Engine references both a Dialect and a Pool, 
which together interpret the DBAPI’s module functions as well as the behavior of the database.)
create_engine(url): Create a new _engine.Engine instance，會回傳Engine，其中參數url可為str或URL.
注意：不同的connection style(如PostgreSQL, MySQL, Oracle, Microsoft SQL Server, SQLite...)有不同的寫法，此處為SQLite寫法。
接著可透過to_sql此method將資料轉為sql檔並將檔案存入engine中作為此db中的一份table。
DataFrame物件名稱.to_sql(name,con): Write records stored in a DataFrame to a SQL database.
Databases supported by SQLAlchemy [1]_ are supported. Tables can be newly created, appended to, or overwritten.
其中參數name代表name of SQL table，注意只能為str；參數con代表sqlalchemy.engine.(Engine or Connection) or sqlite3.Connection
Using SQLAlchemy makes it possible to use any DB supported by that library. Legacy support is provided for sqlite3.Connection 
objects. The user is responsible for engine disposal and connection closure for the SQLAlchemy connectable.
可為str, Connectable或Connection。也可加入參數if_exists代表how to behave if the table already exists，
可為'fail': Raise a ValueError, 'replace': Drop the table before inserting new values
或'append': Insert new values to the existing table，預設為'fail'。'''

from sqlalchemy import create_engine

engine = create_engine("sqlite:///year2021.db")  #建立engine(檔名+附檔名.db)
sortedFrame.to_sql("台灣日本2021",engine,if_exists = "replace")
sortedFrame.to_sql
```

```python
##讀取sql檔
'''可透過pd.read_sql_table(table_name,con)來讀取sql檔：Read SQL database table into a DataFrame.
Given a table name and a SQLAlchemy connectable, returns a DataFrame. This function does not support DBAPI connections.
其中參數table_name代表Name of SQL table in database，注意只能為str；
con代表A database URI could be provided as str. SQLite DBAPI connection mode not supported，可為SQLAlchemy connectable或str。'''

import pandas as pd

dataFrame = pd.read_sql_table("台灣日本2021",engine)
dataFrame.info()
```

```python
##將資料存為json檔
'''可透過DataFrame物件名稱.to_json()此method將資料存為json檔：Convert the object to a JSON string。
可加入參數path_or_buf來指定檔案保存路徑，可為str, path object, file-like object, or None，預設為None；
加入參數orient: Indication of expected JSON string format，可指定json字串格式(指定生成json的key)，
可為"split": dict like {‘index’ -> [index], ‘columns’ -> [columns], ‘data’ -> [values]}, 
"records": list like [{column -> value}, … , {column -> value}], 
"index": dict like {index -> {column -> value}}, 
"columns": dict like {column -> {index -> value}}, 
"values": just the values array，
以及"table": dict like {‘schema’: {schema}, ‘data’: {data}}，預設為"columns"；
參數force_ascii可強制將str編碼為AASCII碼，為bool，預設為True。'''

dataFrame.to_json('台灣日本2021.json',force_ascii = False,orient = "table")
```

```python
##讀取json檔
'''可透過pd.read_json(path_or_buf)此function來讀取json檔：Convert a JSON string to pandas object，戶回傳DataFrame。
其中參數path_or_buf代表愈讀取之json檔案，可為a valid JSON str, path object or file-like object。
另外可加入參數orient: Indication of expected JSON string format指定json字串格式，
注意：若在存為json檔(to_json)時已有指定orient，責此處的orient就要和該指定的orient相同，否則會顯示錯誤。
Compatible JSON strings can be produced by to_json() with a corresponding orient value.'''

import pandas as pd

dataFrame1 = pd.read_json("台灣日本2021.json",orient = "table")
dataFrame1
```
