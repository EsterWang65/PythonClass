```python
#圖表運用--圖表繪製
''' matplotlib為一module: an object-oriented plotting library，而其中matplotlib.pyplot代表繪製圖表部分。
matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. 
It also opens figures on your screen, and acts as the figure GUI manager.
注意：在繪製圖表前要記得先匯入matplotlib.pyplot。
--plt.figure()：Create a new figure, or activate an existing figure.
  此function會建立一圖像(figure)，此語法後的所有圖表皆會繪製在此圖像中，可想為一畫布。
--add_subplot(nrows: int,ncols: int,index:int | tuple[int, int])：Add an ~.axes.Axes to the figure as part of a subplot arrangement. 
  為一method，可在圖像中加入子圖，其中預設有三個參數：
  nrows：幾個rows； nucols：幾個columns； index：代表第幾張圖，為指定創建的axes對象所在的區域，可為整數值或tuple(first,last)。
  系統會按照由左而右、由上而下的順序對每個子區域進行編號，最左上的子區域編號(index)為1。
--plt.plot(x, y, "style_code")可繪製曲線圖，曲線圖plot代表連續的個別資料變化。
--plt.show()：顯示圖像。
--圖表變數名稱.set_title(label:str): Set a title for the Axes.'''

import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure()  #建立圖像(figure)
axes = figure.add_subplot(1,1,1)  #1*1的圖片區域中的(1,1)的位置
x = [0,1,2]  #x座標串列值
y = [3,4,5]  #y座標串列值
axes.plot(x,y,"b--o")  #穿過所有相對應(x, y2)；紅色虛線圓形符號樣式
plt.show()  #顯示圖像
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes1 = figure.add_subplot(1,3,1)
axes1.set_title("title1")
axes2 = figure.add_subplot(1,3,2)
axes2.set_title("title2")
axes3 = figure.add_subplot(1,3,3)
axes3.set_title("title3")

plt.show()
```

```python
#建立多個圖像(figure)
figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
axes1.plot(x,y)

figure2 = plt.figure()
axes2 = figure2.add_subplot(1,1,1)
x1 = [2,3,6]
y1 = [2,1,2]
axes2.plot(x1,y1)

plt.show()
```

```python
#多條曲線在同一圖表中顯示--1
figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
x1 = [2,3,6]
y1 = [2,1,2]
axes1.plot(x,y,'ko-')
axes1.plot(x1,y1,'rD--')
plt.show()
```

```python
#多條曲線在同一圖表中顯示--2
'''.plot()沒有限定引數數量，故上面範例中兩條曲線的語法有可寫在一起。'''
figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
x1 = [2,3,6]
y1 = [2,1,2]
axes1.plot(x,y,'ko-',x1,y1,'rD--')  #黑色實線圓形符號；紅色虛線菱形(diamond)符號
plt.show()
```

```python
#DataFrame, Series及list的運用
import pandas as pd

scores = [
    {'Chinese':78,'English':80,'Math':92,'History':75,'Discover':85},
    {'Chinese':92,'English':60,'Math':85,'History':62,'Discover':73}
]
dataFrame = pd.DataFrame(index=["Robert","Jenny"],data=scores)  #引數名稱呼叫；設定index(列名稱): index or array-like
display(dataFrame)
print(dataFrame.columns)  #DataFrame物件變數名稱.columns為一attribute，會以list-like方式顯示該DataFrame所有的欄位名稱
subjects = list(dataFrame.columns)  #轉型別為list，方便取值
display(subjects)
display(dataFrame["Chinese"])  
#注意：DataFrame實體一般都是用欄位名稱來取值，而非索引編號，且一次只能取一欄，會傳回依只有index沒有欄位名的Series(一維資料)
swapDataFrame = dataFrame.T
display(swapDataFrame)
#DataFrame物件名稱.T為一property，可轉置該DataFrame(index和column對調)。注意：原DataFrame沒有變，該作法是產出一新DataFrame，而非改變
RobertSeries = swapDataFrame["Robert"]
display(RobertSeries)
JennySeries = swapDataFrame["Jenny"]
display(JennySeries)
RobertScores = RobertSeries.to_list()  #Series物件名稱.to_list()為一method，可將該Series的值轉為list，會傳回一list
display(RobertScores)
JennyScores = JennySeries.to_list()
display(JennyScores)

```

```python
#應用：將上面資料轉為圖表-1
'''x軸設定為科目名稱，y軸設定為分數，接著在此圖表上繪製每位學生的分數曲線圖。'''

import matplotlib.pyplot as plt

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,RobertScores,"b-o",label = "Robert")  #設定圖例名稱
axes.plot(subjects,JennyScores,"r-o",label = "Jenny")  #設定圖例名稱
axes.legend()  #圖表物件名稱.legend()：可顯示圖表圖例
plt.show()
```

```python
#應用：將上面資料轉為圖表-2
'''除了上面的方法可繪製圖表外，也可以用pandas.DataFrame裡面的DataFrame名稱.plot()，
系統會自動依據該DataFrame繪製圖表。注意：系統預設會以index作為x軸來繪製圖表，故要先規劃好DataFrame。'''

display(swapDataFrame)
swapDataFrame.plot()
```
