```python
#圖表運用--圖表繪製
''' matplotlib為一module: an object-oriented plotting library，而其中matplotlib.pyplot代表繪製圖表部分。
matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. 
It also opens figures on your screen, and acts as the figure GUI manager.
注意：在繪製圖表前要記得先匯入matplotlib.pyplot。
-- plt.figure() --> Figure: Create a new figure, or activate an existing figure.
   此function會建立一圖像(figure)，此語法後的所有圖表皆會繪製在此圖像中，可想為一畫布。
   可加入參數num: int or str or Figure or SubFigure，可指定該圖像/畫布的標記(unique identifier)。
   預設為None，若不指定，每次執行figure()時會自動增加1，表示會產生新的圖像/畫布；
   也可加入參數figsize(float, float)來指定畫布尺寸(Width, Height)，單位為英吋。
-- .add_subplot(nrows: int, ncols: int, index:int | tuple[int, int])  --> Axes: Add an ~.axes.Axes to the figure as part of a subplot arrangement. 
   為一method，可在圖像中加入子圖，會傳回Axes實體，其中預設有三個參數：
   nrows：幾列rows； ncols：幾欄columns； index：代表第幾張圖，為指定創建的axes對象所在的區域，可為整數值或tuple(first, last)，
   系統會按照由左而右、由上而下的順序對每個子區域進行編號，最左上的子區域編號(index)為1。
-- .plot(x, y, "style_code")  --> list of Line2D(A list of lines representing the plotted data.): Plot y versus x as lines and/or markers. 
   為一method，可繪製曲線圖，曲線圖plot代表連續的個別資料變化。其中參數"style_code"即format string，可指定輸出格式，注意只能為str。 
   A format string consists of a part for color, marker and line: fmt = '[marker][line][color]'(順序可變動).
-- plt.show()  --> None: Display all open figures. 為一function，顯示圖像。
-- 圖表變數名稱.set_title(label: str)  --> Text: Set a title for the Axes. 為一method。'''

import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure()  #建立圖像(figure)
axes = figure.add_subplot(1,1,1)  #1*1的圖片區域中的(1,1)的位置
x = [0,1,2]  #x座標串列值
y = [3,4,5]  #y座標串列值
axes.plot(x,y,"b--o")  
#穿過所有相對應的(x,y)座標；紅色虛線圓形符號樣式。注意：x及y值數量需相等，否則會顯示錯誤ValueError: x and y must have same first dimension, but have shapes (3,) and (4,)
plt.show()  #顯示圖像
```

```python
#建立多個圖像(figure)
figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
axes1.plot(x,y)

figure2 = plt.figure()
axes2 = figure2.add_subplot(1,1,1)
x1 = [2,3,6]
y1 = [2,1,2]
axes2.plot(x1,y1)

plt.show()
```

```python
#多條曲線在同一圖表中顯示--1
figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
x1 = [2,3,6]
y1 = [2,1,2]
axes1.plot(x,y,'ko-')  #黑色實線圓形符號
axes1.plot(x1,y1,'rD--')  #紅色虛線菱形(diamond)符號
plt.show()
```

```python
#多條曲線在同一圖表中顯示--2
'''.plot()沒有限定引數數量，故上面範例中兩條曲線的語法有可寫在一起。'''
figure1 = plt.figure()
axes1 = figure1.add_subplot(1,1,1)
x = [0,1,2]
y = [3,4,5]
x1 = [2,3,6]
y1 = [2,1,2]
axes1.plot(x,y,'ko-',x1,y1,'rD--')  #沒有限定引數數量
plt.show()
```

```python
#DataFrame, Series及list的運用

import pandas as pd

##建立一裡面放dict的list
scores = [
    {'Chinese':78,'English':80,'Math':92,'History':75,'Discover':85},
    {'Chinese':92,'English':60,'Math':85,'History':62,'Discover':73}
]
dataFrame = pd.DataFrame(index = ["Robert","Jenny"], data = scores)  #引數名稱呼叫；設定index(列名稱)：index or array-like
display(dataFrame)

##DataFrame轉型別&取值
print(dataFrame.columns)  #DataFrame物件變數名稱.columns為一attribute，會以list-like(可容易轉為list)方式顯示該DataFrame所有的欄位名稱
subjects = list(dataFrame.columns)  #轉型別為list，方便取值
display(subjects)
display(dataFrame["Chinese"])  #注意：DataFrame實體一般都是用欄位名稱來取值，而非索引編號，且一次只能取一欄，會傳回一只有index沒有欄位名的Series(一維資料)

##DataFrame轉置&取值
swapDataFrame = dataFrame.T  #DataFrame物件名稱.T為一property，可轉置該DataFrame(index和column對調)。注意：原DataFrame沒有變，該作法是產出一新DataFrame，而非改變
display(swapDataFrame)
RobertSeries = swapDataFrame["Robert"]
display(RobertSeries)
JennySeries = swapDataFrame["Jenny"]
display(JennySeries)

##將Series轉為list
RobertScores = RobertSeries.to_list()  #Series物件名稱.to_list() --> list: 為一method，可將該Series的值轉為list，會傳回一list
display(RobertScores)
JennyScores = JennySeries.to_list()
display(JennyScores)

```

```python
#應用：將上面資料轉為圖表-1
'''x軸設定為科目名稱，y軸設定為分數，接著在此圖表上繪製每位學生的分數曲線圖。'''

import matplotlib.pyplot as plt

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(subjects, RobertScores, "b-o", label = "Robert")  #設定圖例名稱
axes.plot(subjects, JennyScores, "r-o", label = "Jenny")  #設定圖例名稱
axes.legend()  #圖表物件名稱.legend() --> Legend：可顯示圖表圖例
plt.show()
```

```python
#應用：將上面資料轉為圖表-2
'''除了上面的方法可繪製圖表外，也可以用pandas.DataFrame裡面的property: DataFrame名稱.plot()，
系統會自動依據該DataFrame繪製圖表。注意：系統預設會以index作為x軸來繪製圖表，故要先規劃好DataFrame。
可直接寫swapDataFrame.plot()即可顯示圖表，但若如以下匯入模組工具來運用即可增加對圖表的操控靈活度。
另外也可透過在DataFrame.plot()內加入參數來操控圖表繪製，如style: list or dict(若有兩筆資料以上方便設定)可設定曲線樣式；
marker可指定符號樣式；kind: str可指定圖表類型('line', 'bar', 'box'...)等。
pandas.DataFrame.plot(): Make plots of Series or DataFrame. '''

import matplotlib.pyplot as plt

display(swapDataFrame)
figure = plt.figure()
axes = figure.add_subplot(1, 1, 1)
swapDataFrame.plot(ax = axes, style = ["--", '-'], marker = 'o')  
#加入參數ax可設定圖表要繪製在哪個figure的哪個位置，預設為None。ax = matplotlib axes object: an axes of the current figure；第一/二筆資料為虛線/實線
axes.set_title("Scores")
axes.grid(visible = True,axis = 'y')  
#matplotlib.pyplot.grid(): Configure the grid lines. 為一method，將參數visible設為True即可產生格線，可加入參數axis = 'x'/'y'/"both"來指定要顯示的格線
plt.show()
```

```python
#中文處理
'''若圖表中有中文字則要記得先灌字型，如此才能顯示中文字。
注意：matplotlib該module中的matplotlib.font_manager代表圖表的字型管理，記得要先匯入。
注意：要先到終端機灌字型，須分別執行以下兩行命令：
sudo apt-get update -y  以及
sudo apt-get install -y fonts-wqy-zenhei  '''

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname = '/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size = 12)  #codespace中文字處理

dataFrame1 = pd.DataFrame(index=["小明","小華"],data=scores)
swapDataFrame1 = dataFrame1.T
display(swapDataFrame1)
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
swapDataFrame1.plot(ax = axes,style = ["--",'-'],marker = 'o')  
axes.set_title("成績", fontproperties = font)  #想顯示中文字需加入參數fontproperties，並將其指定給預先灌好的字型
axes.grid(visible = True,axis = 'y')
plt.legend(prop = font)
plt.show()
```
